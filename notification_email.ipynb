```python
import smtplib
from email.mime.text import MIMEText
from datetime import datetime, timedelta
import pandas as pd
import re 

# Set up your email server and credentials
smtp_port = 587
smtp_server = "smtp.office365.com"
smtp_username = "your_username"
smtp_password = "your_password"

# User Input
product_name_input = input("Enter the product name: ")
expiration_date_input = input("Enter the expiration date (YYYY-MM-DD): ")

# Merge DataFrames based on 'name_id' with a full outer join
merged_df = pd.merge(smart_pantry, recipes_df, on = "name_id", how = "left")

# Convert expiration_date_input to a datetime object
expiration_date_input = datetime.strptime(expiration_date_input, "%Y-%m-%d")

# Generate the email content with all products and associated recipes
email_content = ""
for index, row in merged_df.iterrows():
    product_name = row["products"]
    recipe_title = row["recipe_title"]
    preparation_time = row["preparation_time"]
    recipe_url = row["recipe_url"]
    instructions = row["instructions"]
    ingredients_list = row["ingredients_list"]

    email_content += f"Product: {product_name}\n"
    email_content += f"Recipe Title: {recipe_title}\n"
    email_content += f"Preparation Time: {preparation_time}\n"
    email_content += f"Recipe URL: {recipe_url}\n"
    email_content += f"Instructions: {instructions}\n"
    email_content += f"Ingredients List: {ingredients_list}\n\n"

# Create the email message
subject = f"Recipes for {product_name_input}"
message = f"Hello,\n\nHere are some recipes for {product_name_input}:\n\n{email_content}\nBest regards,\nYour Smart Pantry"

msg = MIMEText(message)
msg["Subject"] = subject
msg["From"] = "your_email"
msg["To"] = "other_email"

# Send the email
with smtplib.SMTP(smtp_server, smtp_port) as server:
    server.starttls()
    server.login(smtp_username, smtp_password)
    server.sendmail("your_email", ["other_email"], msg.as_string())
```
